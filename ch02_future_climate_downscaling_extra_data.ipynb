```python
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd
import numpy as np
import cartopy.crs as ccrs
import palettable
import cartopy.feature as cfeature
```

## Maximum daily temperature (`tasmax`)

```python
! gdown https://drive.google.com/file/d/16xSA1HbMIqtDq7bP31O_LfkwiVdXNNgX/view?usp=drive_link --fuzzy -O tasmax_EC-Earth3-Veg_1990-2050.nc
```

```python
var = 'tasmax'

ds = xr.open_dataset(f"tasmax_EC-Earth3-Veg_1990-2050.nc")
fig, ax = plt.subplots(subplot_kw={"projection": ccrs.PlateCarree()})

# Plot
ds[var].isel(time=0).plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    x="lon",
    y="lat",
    cmap=palettable.colorbrewer.diverging.RdYlBu_10_r.mpl_colormap,
)

# Add coastlines
ax.coastlines(resolution='50m')

# Set the title
ax.set_title("")

# Show the plot
plt.show()
```
